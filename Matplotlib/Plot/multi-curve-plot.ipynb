```python
import matplotlib.pyplot as plt
```

```python
x = [num for num in range(0,99)]
```

##### y1= x* 100
##### y2= x ^ 2
##### y3= 0

```python
y1 = list(map(lambda num: num*100, x))
y2 = list(map(lambda num: num ** 2, x))
y3 = [0 for numbers in x]
```

```python
plt.plot(x, y1, c="red")
plt.plot(x, y2, c="blue")
plt.plot(x, y3, c="purple")

plt.xlabel('Number')
plt.ylabel('calculation')
plt.title('show multiple curves')
plt.show()
```
